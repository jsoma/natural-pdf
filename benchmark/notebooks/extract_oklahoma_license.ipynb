```python
from natural_pdf import PDF

pdf = PDF("../pdfs/m27.pdf")
page = pdf.pages[1]
# page.show()
```

```python
# Explore the page
headers = page.region(top=55, height=30).find_all('text').expand(1).dissolve().expand(-1)
headers.show()
```

```python
region = page.find("text:contains(LICENSEE NAME)").below(until='text:regex(Page \d)', include_endpoint=False)
```

```python
region.show()
```

```python
rows = headers[0].below(width='element').find_all('text', overlap='partial')
rows.show(crop=20)
```

```python
region.extract_table(    
    verticals=headers,   
    horizontals=rows.map(lambda r: r.bbox[3]),                                                                             
    outer=True                                                                                         
).to_df(header=headers)  # Headers auto-sorted by x0                                                   
```
